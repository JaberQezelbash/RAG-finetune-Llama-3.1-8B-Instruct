# Token Access from HuggingFace

```python
# 00_token_login.py
# Run this first in Jupyter or as a Python script.
# It asks for your Hugging Face token, logs in, and verifies Llama access.

import os
import getpass

from huggingface_hub import login
from transformers import AutoTokenizer


def main() -> None:
    hf_token = getpass.getpass("Paste your Hugging Face token. It will not be shown: ").strip()

    if not hf_token:
        raise RuntimeError("No Hugging Face token was entered.")

    os.environ["HF_TOKEN"] = hf_token
    login(token=hf_token, add_to_git_credential=False)

    print("Token saved for this session.")

    tokenizer = AutoTokenizer.from_pretrained(
        "meta-llama/Llama-3.1-8B-Instruct",
        token=os.environ["HF_TOKEN"],
    )

    print("Llama access works.")


if __name__ == "__main__":
    main()
```
